```python
import pandas as pd
```

```python
bloomberg_not_include_score_processed = pd.read_csv('../CSR_score/bloomberg_not_include_score_processed.csv')
refinitiv_ESG_score = pd.read_csv('../CSR_score/refinitiv_ESG_score_100.csv')
```

```python
bloomberg_not_include_score_processed.head()
```

```python
refinitiv_ESG_score.head()
```

```python
common_years = [str(year) for year in range(2005, 2025)]
refinitiv_ESG_score_filtered = refinitiv_ESG_score[['ticker'] + common_years]

merged = bloomberg_not_include_score_processed.merge(refinitiv_ESG_score_filtered, on='ticker', suffixes=('_bloomberg', '_refinitiv'))

for year in common_years:
    col_bloomberg = f"{year}_bloomberg"
    col_refinitiv = f"{year}_refinitiv"

    merged[year] = merged.apply(lambda row: 0 if row[col_bloomberg] == 0 or row[col_bloomberg] == -1 or 
                        row[col_refinitiv] == 0 or row[col_refinitiv] == -1 
                    else row[col_bloomberg] - row[col_refinitiv], axis=1)
    
merged = merged[['ticker'] + common_years]
merged.to_csv('../CSR_score/common_score_ESG_0319.csv', index=False)
```

```python
common_score_ESG_0319 = pd.read_csv('../CSR_score/common_score_ESG_0319.csv')
common_score_ESG = pd.read_csv('../CSR_score/common_score_ESG.csv')
```

```python
common_score_ESG_all = pd.concat([common_score_ESG, common_score_ESG_0319], ignore_index=True)
common_score_ESG_all.to_csv('../CSR_score/common_score_ESG_all.csv', index=False)
```
