```python
def divide_folder_name(dataset):
    dataset['exchange'] = dataset['folder_name'].str.split('_').str[0]
    dataset['ticker'] = dataset['folder_name'].str.split('_').str[1]
    dataset['date'] = dataset['folder_name'].str.split('_').str[2]
    return dataset
```
